```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Dados do CFD
data = {
    'W1': [4, 2, 0],
    'W2': [17, 10, 7],
    'W3': [30, 20, 11],
    'W4': [43, 33, 19],
    'W5': [56, 43, 29],
    'W6': [69, 51, 40],
    'W7': [82, 61, 46],
    'W8': [93, 72, 54],
    'W9': [102, 83, 56],
    'W10': [110, 93, 56]
}
data
```

```python
# Criação de um DataFrame do pandas
df = pd.DataFrame(data, index=['Novos Incidentes', 'Em Execução', 'Fechados'])
df
```

```python
# Criação do gráfico CFD
plt.figure(figsize=(12, 6))
plt.stackplot(range(1, 11), df.loc['Fechados'], df.loc['Em Execução'], df.loc['Novos Incidentes'],
              labels=['Fechados', 'Em Execução', 'Novos Incidentes'], colors=['#00ff00', '#ff9900', '#3399ff'])
plt.title('Cumulative Flow Diagram (CFD)')
plt.xlabel('Semanas (W)')
plt.ylabel('Número de Incidentes')
plt.legend(loc='upper left')
plt.margins(0, 0)  # Remove as margens para melhor visualização
plt.xticks(range(1, 11), ['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8', 'W9', 'W10'])
plt.grid(True)

# Exibição do gráfico CFD
plt.show()
```
